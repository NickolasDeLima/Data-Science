
# Logistic Regression

Para este projeto, o Logistic Regression sera usado para prever se um consumidor irá clicar ou não em um anúncio.

---

# Dados

   - **'Daily Time Spent on Site'**: tempo do consumidor no site (minutos).
   - **'Age'**: idade do cliente em anos.
   - **'Area Income'**: Renda média da área geográfica do consumidor.
   - **'Daily Internet Usage'**: Média de minutos por dia que o consumidor está na internet.
   - **'Ad Topic Line'**: Título do anúncio.
   - **'City'**: Cidade do consumidor.
   - **'Male'**: Se o consumidor era ou não do sexo masculino.
   - **'Country'**: País do consumidor.
   - **'Timestamp'**: Momento em que o consumidor clicou no anúncio ou fechou a janela.
   - **'Clicked on Ad'**: 0 ou 1 indica se clicou ou não no anúncio.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from pandas_profiling import ProfileReport
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('../input/my-datasets/advertising.csv')
df.head()
```

```python
report = ProfileReport(df)
report
```

## Logistic Regression

```python
X = df[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = df['Clicked on Ad']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

## Previsões e avaliações

```python
predictions = logmodel.predict(X_test)

print(classification_report(y_test,predictions))
```
